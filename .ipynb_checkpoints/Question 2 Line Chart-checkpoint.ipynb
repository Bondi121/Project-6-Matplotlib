```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

videogames = pd.read_csv('top100videogamesales.csv')
videogames.head(10)
```

```python
videogames_grouped = videogames.groupby('Publisher').sum()
videogames_grouped = videogames_grouped.sort_values('Global_Sales')
videogames_grouped
```

```python
videogames_grouped = videogames.groupby('Publisher').count()
videogames_grouped = videogames_grouped.sort_values('Global_Sales')
videogames_grouped
```

```python
videogames_grouped = videogames.groupby('Global_Sales').sum()
videogames_grouped = videogames_grouped.sort_values('Global_Sales')
videogames_grouped
```

```python
grouped_by_Genre = videogames.groupby('Genre')
grouped_by_Genre.size().sort_values(ascending=False)[:5]
```

```python
grouped_by_year = videogames.groupby('Year').sum()
grouped_by_year
```

```python
grouped_by_year = videogames.groupby('Year').count()
grouped_by_year
```

## Question 2: Global Sales by Year: Create a line chart to show the total global sales for each year. Explain the result of the chart in markdow

```python
yearly_global_sales = videogames[['Year', 'Global_Sales']].groupby('Year').sum()
yearly_global_sales.plot(kind='line', figsize=(12, 6), color="green", linestyle="dotted", title = 'Total Global Sales by the year')
plt.ylabel("Sales (Mio Dollars")
```

```python
global_sales_by_year = videogames[['Year', 'Global_Sales']].groupby('Year').sum()
global_sales_by_year
```
